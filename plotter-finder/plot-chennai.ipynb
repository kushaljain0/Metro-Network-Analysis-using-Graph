```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import folium
```

```python
df = pd.read_csv('../complete-dataset/chennai-metro-data.csv')
```

```python
df.head()
```

```python
avglatitude = df['Latitude'].mean()
avglongitude = df['Longitude'].mean()
print("Average Latitude : ", avglatitude)
print("Average Longitude : ", avglongitude)
```

```python
map = folium.Map(location=[avglatitude, avglongitude], zoom_start=10)

tooltip = "Click Here For More Info"

def add_point(name, lat, lon, map):
    marker = folium.CircleMarker(
        location = [lat,lon],
        popup = f"<stong>{name} Metro Sation</stong>",
        radius = 1, weight = 2)
    marker.add_to(map)

for index, row in df.iterrows():
    add_point(row['Station Name'], row['Latitude'], row['Longitude'], map)
    
map
```
